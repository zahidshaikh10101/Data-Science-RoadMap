```python
# List of employee dictionaries
employees = [
    {"name": "Alice", "scores": (78, 82, 90), "active": "True"},
    {"name": "Bob", "scores": (60, 55, 58), "active": "False"},
    {"name": "Charlie", "scores": (88, 91, 94), "active": "True"},
    {"name": "Diana", "scores": (45, 50, 47), "active": "True"},
    {"name": "Evan",   "scores": (70, 65, 68), "active": "False"}
]

# Function to calculate average of a tuple of scores
def calculate_average(scores):
    return sum(scores) / len(scores)

# Step 1: Convert 'active' string to boolean, calculate average, and assign category
for emp in employees:
    # Convert active from string to boolean
    emp['active'] = emp['active'] == "True"

    # Calculate average score
    emp['avg'] = calculate_average(emp['scores'])

    # Assign category based on average
    if emp['avg'] >= 85:
        emp['category'] = "Excellent"
    elif emp['avg'] >= 70:
        emp['category'] = "Good"
    elif emp['avg'] >= 50:
        emp['category'] = "Average"
    else:
        emp['category'] = "Poor"

# Step 2: Group active employees by category
grouped = {}

for emp in employees:
    # Warning for very low scores
    if min(emp['scores']) < 40:
        print("Warning: Very low score detected!")
        break

    # Skip inactive employees
    if not emp['active']:
        continue

    category = emp['category']
    name = emp['name']

    # Add employee to grouped dictionary
    if category not in grouped:
        grouped[category] = []
    grouped[category].append(name)

# Step 3: Count total active employees processed
active_count = sum(1 for emp in employees if emp['active'])

# Step 4: Print results
print("Updated Employee List:")
for emp in employees:
    print(emp)

print("\nGrouped by Category (Active Employees Only):")
print(grouped)

print("\nTotal Active Employees Processed:", active_count)
```

# Recursion

```python
def factorial(x):
  if x == 0 or x == 1:
    return 1
  else:
    return x * factorial(x - 1)

print(factorial(5))
```

```python
def fibbo(x):
  if x == 0 or x == 1:
    return 1
  else:
    return fibbo(x-1) + fibbo(x-2)

fibbo(7)
```

# Decorators

```python
def add_sprinkles(func):
  def wrapper(*args, **kwargs):
    print("Adding Sprinkles")
    func(*args, **kwargs)
  return wrapper

def add_fudge(func):
  def wrapper(*args, **kwargs):
    print("Adding Fudge")
    func(*args, **kwargs)
  return wrapper

def get_ice_cream(flavor):
  print(f"Here is your {flavor} icecream.")

final = add_sprinkles(add_fudge(get_ice_cream))
final("Chocolate")
```

# Iterators & Generators

```python
class TopTen:

  def __init__(self):
    self.num = 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.num <= 10:
      val = self.num
      self.num += 1
      return val

    else:
      raise StopIteration


values = TopTen()

for i in values:
  print(i)
```

1. Write a recursive function to generate all subsets (power set) of a given list. - Hard

```python
listin = [1, 2, 3, 4]

def subset(x):
  emp = [[]]
  for i in x:
    emp.append(i)
  for i in range(len(x)):
    for j in range(i+1, len(x)):
      emp.append([x[i], x[j]])

  for i in range(len(x)):
    for j in range(i+1, len(x)):
      for k in range(j+1, len(x)):
        emp.append([x[i], x[j], x[k]])
  emp.append(x)
  return emp

subset(listin)
```

```python
def greet(func):
  def wrapper(*args, **kwargs):
    print("Hello")
    result = func(*args, **kwargs)
    print("Welcome")
    return result
  return wrapper

def power_set(nums):
    if len(nums) == 0:
        return [[]]
    first = nums[0]
    rest = nums[1:]
    subsets_without_first = power_set(rest)
    subsets_with_first = []
    for subset in subsets_without_first:
        subsets_with_first.append([first] + subset)

    return subsets_without_first + subsets_with_first

input  = [1, 2, 3]
result = greet(power_set)
print(result(input))
```
